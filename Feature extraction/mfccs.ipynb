# MEL FREQUENCY CEPSTRAL COEFFICIENTS (MFCCs)

The Mel frequency cepstral coefficients (MFCCs) of a signal are a small set of features (usually about 10â€“20) which concisely describe the overall shape of a spectral envelope.
'librosa.feature.mfcc' computes MFCCs across an audio signal.

```python
import librosa as lr
import librosa.display
import matplotlib.pyplot as plt
import sklearn.preprocessing 

import matplotlib.style as ms
%matplotlib inline
import numpy as np
import os
```

```python
audio='/home/vivek/Genre_Detection/genres/blues/blues.00000.au'
x , sr =lr.load(audio)
```

```python
mfccs = lr.feature.mfcc(x, sr=sr)
mfccs.shape
```

The shape of the numpy array returned is (20,1293) i.e. it computed 20 MFCCs over 1293 frames (all the features extracted are computed over 1293 frames for given 30 s audio)
MFCCs are basically collection of 20 features which can help in classifying the genre whem model is trained over them

```python
#DISPLAYING MFCCs
plt.figure(figsize=(12,4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```

```python
for e in mfccs :
    print(np.mean(e))
```

These 20 values are used in the genre determination
